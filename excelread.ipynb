```python
for i in range(1,21):
    temp_row = sheet1.row_values(i)
    str_out = '\"'+str(temp_row[0])+"\", "
    for j in range(1,16):
        if j == 4 or j == 5:
            str_out += str(eval(temp_row[j]))+", "
        elif j == 9:
            str_out += str(temp_row[j].split('+')[0])+", "
        elif j == 14:
            if temp_row[j] == 'n':
                str_out += "0, "
            else:
                str_out += "1, "
        else:
            str_out +=  str(temp_row[j])+", " 
    print(str_out)
    
```
